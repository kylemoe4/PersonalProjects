```python
def reverseDict(d):
    replacement = {'FGM':'FGM-','FGA':'FGA-','FG_PCT':'FG_PCT-','FG3M':'FG3M-',
                   'FG3A':'FG3A-','FG3_PCT':'FG3_PCT-','FTM':'FTM-','FTA':'FTA-',
                   'FT_PCT':'FT_PCT-','OREB':'OREB-','DREB':'DREB-','REB':'REB-',
                   'AST':'AST-','TOV':'TOV-','STL':'STL-','BLK':'BLK-',
                   'BLKA':'BLKA-','PF':'PF-','PFD':'PFD-','WPCT':'WPCT-',
                   'FGM-':'FGM','FGA-':'FGA','FG_PCT-':'FG_PCT','FG3M-':'FG3M',
                   'FG3A-':'FG3A','FG3_PCT-':'FG3_PCT','FTM-':'FTM','FTA-':'FTA',
                   'FT_PCT-':'FT_PCT','OREB-':'OREB','DREB-':'DREB','REB-':'REB',
                   'AST-':'AST','TOV-':'TOV','STL-':'STL','BLK-':'BLK',
                   'BLKA-':'BLKA','PF-':'PF','PFD-':'PFD','WPCT-':'WPCT'}
    return pd.Series({replacement[k]:v for k, v in d.items()})
```

```python
def predictor(team, gameDate):
    teamGames = [games[0][games[0]['TEAM_ABBREVIATION']==team].copy(),
                 games[1][games[1]['TEAM_ABBREVIATION']==team].copy(),
                 games[2][games[2]['TEAM_ABBREVIATION']==team].copy(),
                 games[3][games[3]['TEAM_ABBREVIATION']==team].copy(),
                 games[4][games[4]['TEAM_ABBREVIATION']==team].copy(),
                 games[5][games[5]['TEAM_ABBREVIATION']==team].copy(),
                 games[6][games[6]['TEAM_ABBREVIATION']==team].copy()]
    for i in range(len(teamGames)):
        teamGames[i]['GAME_DATE'] = teamGames[i]['GAME_DATE'].apply(lambda x: x[:10])
        teamGames[i] = teamGames[i][teamGames[i]['GAME_DATE'] < gameDate]
    teamID = teamGames[0].iloc[0]['TEAM_ID']
    # get current season from date
    year = 0
    for i in range(len(teamGames)):
        if (len(teamGames[i]) == 0):
            year = i-1
            break
        year = i
    wprev = 0
    wcurr = 0
    if (len(teamGames[year]) < 5):
        wprev = 0.2*(5-len(teamGames[year]))
        wcurr = 0.2*len(teamGames[year])
    else: 
        wcurr = 1
        teamGames[year] = teamGames[year][-5:]
    avs = teamGames[year][['FGM','FGA','FG_PCT','FG3M','FG3A','FG3_PCT','FTM','FTA','FT_PCT','OREB','DREB','REB','AST','TOV','STL','BLK','BLKA','PF','PFD','WPCT','FGM-','FGA-','FG_PCT-','FG3M-','FG3A-','FG3_PCT-','FTM-','FTA-','FT_PCT-','OREB-','DREB-','REB-','AST-','TOV-','STL-','BLK-','BLKA-','PF-','PFD-','WPCT-']].mean()
    for i, v in avs.items():
        if (i != 'WPCT'):
            avs = avs.replace(v, wcurr*v + wprev*m[year][0][0][i])
        else:
            avs = avs.replace(v, wcurr*v + wprev*records[year-1][teamID])
    return avs
    
```
