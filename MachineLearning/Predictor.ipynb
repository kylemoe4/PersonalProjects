```python
import numpy as np
import pandas as pd
import math
import sklearn
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
data = pd.read_csv(r'./data/SteamboatDataScaled.csv', index_col=0)
```

```python
dataKNN = data.drop(columns=['HOUR0','HOUR1','HOUR2','HOUR3','HOUR4','HOUR5','HOUR6','HOUR7','HOUR8','HOUR9','HOUR10','HOUR11','HOUR12','HOUR13','HOUR14','HOUR15','HOUR16','HOUR17','HOUR18','HOUR19','HOUR20','HOUR21','HOUR22','HOUR23','DATE'])
```

```python
from sklearn.neighbors import NearestNeighbors
nn = NearestNeighbors(n_neighbors=11).fit(dataKNN)
nearest = nn.kneighbors(dataKNN)
```

```python
from sklearn.mixture import GaussianMixture
```

```python
gauss = GaussianMixture(n_components=30).fit(data.drop(columns="DATE"))
```

```python
y_true = gauss.predict(data.drop(columns="DATE"))
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, y_true, test_size=0.2, random_state=42)
```

```python
X_test.head()
```

```python
y_true = gauss.predict(X_test.drop(columns="DATE"))
```

```python
X_trainKNN = X_train.drop(columns=['HOUR0','HOUR1','HOUR2','HOUR3','HOUR4','HOUR5','HOUR6','HOUR7','HOUR8','HOUR9','HOUR10','HOUR11','HOUR12','HOUR13','HOUR14','HOUR15','HOUR16','HOUR17','HOUR18','HOUR19','HOUR20','HOUR21','HOUR22','HOUR23','Flux','DATE'])
X_testKNN = X_test.drop(columns=['HOUR0','HOUR1','HOUR2','HOUR3','HOUR4','HOUR5','HOUR6','HOUR7','HOUR8','HOUR9','HOUR10','HOUR11','HOUR12','HOUR13','HOUR14','HOUR15','HOUR16','HOUR17','HOUR18','HOUR19','HOUR20','HOUR21','HOUR22','HOUR23','Flux','DATE'])
nn = NearestNeighbors(n_neighbors=10).fit(X_trainKNN)
nearest = nn.kneighbors(X_testKNN)
```

```python
X_testKNN.head()
```

```python
# randomize TMAX, TMIN, SNOW
```

```python
indices = X_test.index.values.tolist()

near = {indices[i]: nearest[1][i].tolist() for i in range(len(indices))}
#print(near)
```

```python
for i in near:
    aveTraffic = X_train.iloc[near[i]][['HOUR0','HOUR1','HOUR2','HOUR3','HOUR4','HOUR5','HOUR6','HOUR7','HOUR8','HOUR9','HOUR10','HOUR11','HOUR12','HOUR13','HOUR14','HOUR15','HOUR16','HOUR17','HOUR18','HOUR19','HOUR20','HOUR21','HOUR22','HOUR23']].mean()
    X_test.loc[i,['HOUR0','HOUR1','HOUR2','HOUR3','HOUR4','HOUR5','HOUR6','HOUR7','HOUR8','HOUR9','HOUR10','HOUR11','HOUR12','HOUR13','HOUR14','HOUR15','HOUR16','HOUR17','HOUR18','HOUR19','HOUR20','HOUR21','HOUR22','HOUR23']] = aveTraffic
X_test['Flux'] = X_test[['HOUR0','HOUR1','HOUR2','HOUR3','HOUR4','HOUR5','HOUR6','HOUR7','HOUR8','HOUR9','HOUR10','HOUR11','HOUR12','HOUR13','HOUR14','HOUR15','HOUR16','HOUR17','HOUR18','HOUR19','HOUR20','HOUR21','HOUR22','HOUR23']].sum(axis=1)
X_test.head()
```

```python
y_hat_proba = gauss.predict_proba(X_test.drop(columns=['DATE']))
```

```python
y_hat_first = []
y_hat_second = []
y_hat_third = []
for i in range(len(y_hat_proba)):
    y_hat_first.append(np.argpartition(y_hat_proba[i], -2)[-1:][0])
    y_hat_second.append(np.argpartition(y_hat_proba[i], -2)[-2:-1][0])
    y_hat_third.append(np.argpartition(y_hat_proba[i], -2)[-3:-2][0])
```

```python
import sklearn.metrics
```

```python
print(sklearn.metrics.accuracy_score(y_true, y_hat_first))
```

```python
print(sklearn.metrics.accuracy_score(y_true, y_hat_second))
```

```python
print(sklearn.metrics.accuracy_score(y_true, y_hat_third))
```
