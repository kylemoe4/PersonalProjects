```python
import numpy as np
import pandas as pd
import math
import sklearn
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.mixture import GaussianMixture
from sklearn.cluster import DBSCAN
data = pd.read_csv(r'./data/SteamboatDataScaled.csv', index_col=0)
unscaledData = pd.read_csv(r'./data/SteamboatData.csv', index_col=0)
unscaledAgg = pd.read_csv(r'./data/SteamboatData.csv', index_col=0)
unscaledGaussian = pd.read_csv(r'./data/SteamboatData.csv', index_col=0)
unscaledDBSCAN = pd.read_csv(r'./data/SteamboatData.csv', index_col=0)

data.head()
```

```python
#data[['MON','TUE','WED','THU','FRI','SAT','SUN']] = data[['MON','TUE','WED','THU','FRI','SAT','SUN']]/2
```

```python
for i in range(30,2,-1):
    kmeans = KMeans(n_clusters=i).fit(data.drop(columns="DATE"))
    unscaledData['KMEANS' + str(i)] = kmeans.labels_
```

```python
unscaledData.to_csv('./data/KMeansResults.csv')
```

```python
for i in range(30,2,-1):
    agg = AgglomerativeClustering(n_clusters=i,linkage='complete').fit(data.drop(columns="DATE"))
    unscaledAgg['AGG' + str(i)] = agg.labels_
```

```python
unscaledAgg.to_csv('./data/AgglomerativeClusteringResults.csv')
```

```python
for i in range(30,2,-1):
    gauss = GaussianMixture(n_components=i).fit(data.drop(columns="DATE"))
    unscaledGaussian['GAUSS' + str(i)] = gauss.predict(data.drop(columns="DATE"))
```

```python
unscaledGaussian.to_csv('./data/GaussianMixtureResults.csv')
```

```python
dbscan = DBSCAN(eps=.5).fit(data.drop(columns=["DATE"]))
unscaledDBSCAN['DB'] = dbscan.labels_
```

```python
unscaledDBSCAN.to_csv('./data/DBSCANResults.csv')
```
