```python
col_list = ['HOUR0','HOUR1','HOUR2','HOUR3','HOUR4','HOUR5','HOUR6','HOUR7','HOUR8','HOUR9','HOUR10','HOUR11','HOUR12','HOUR13','HOUR14','HOUR15','HOUR16','HOUR17','HOUR18','HOUR19','HOUR20','HOUR21','HOUR22','HOUR23']
sn['Flux'] = sn[col_list].sum(axis=1)
sn['PrevFlux24'] = sn['Flux'].rolling(window=1, min_periods=1).sum().shift().bfill()
sn['PrevFlux48'] = sn['Flux'].rolling(window=2, min_periods=1).sum().shift().bfill()
```

```python
skiing = weather.merge(sn, how='inner', left_on='DATE', right_on='FormattedDate').fillna(0)
```

```python
skiing = skiing.drop(columns=['STATION','NAME','FormattedDate','TOBS'])
```

```python
skiing['SNOW24'] = skiing['SNOW'].rolling(window=1, min_periods=1).sum().shift().bfill().round(decimals=1)
skiing['SNOW48'] = skiing['SNOW'].rolling(window=2, min_periods=1).sum().shift().bfill().round(decimals=1)
skiing['SNOW7'] = skiing['SNOW'].rolling(window=7, min_periods=1).sum().shift().bfill().round(decimals=1)
```

```python
skiing = skiing[~skiing['DATE'].dt.month.isin([5,6,7,8,9,10])]
skiing.head()
```

```python
# save to csv file

skiing.to_csv('./data/SteamboatData.csv')
```

```python
skiing = skiing.drop(columns=['PRCP'])
```

```python
skiing['MON'] = np.where(skiing['WEEKDAY'] == 0, 1, 0)
skiing['MON'] = np.where((skiing['WEEKDAY'] == 1) | (skiing['WEEKDAY'] == 6), 0.5,skiing['MON'])
skiing['TUE'] = np.where(skiing['WEEKDAY'] == 1, 1, 0)
skiing['TUE'] = np.where((skiing['WEEKDAY'] == 2) | (skiing['WEEKDAY'] == 0), 0.5,skiing['TUE'])
skiing['WED'] = np.where(skiing['WEEKDAY'] == 2, 1, 0)
skiing['WED'] = np.where((skiing['WEEKDAY'] == 3) | (skiing['WEEKDAY'] == 1), 0.5,skiing['WED'])
skiing['THU'] = np.where(skiing['WEEKDAY'] == 3, 1, 0)
skiing['THU'] = np.where((skiing['WEEKDAY'] == 4) | (skiing['WEEKDAY'] == 2), 0.5,skiing['THU'])
skiing['FRI'] = np.where(skiing['WEEKDAY'] == 4, 1, 0)
skiing['FRI'] = np.where((skiing['WEEKDAY'] == 5) | (skiing['WEEKDAY'] == 3), 0.5,skiing['FRI'])
skiing['SAT'] = np.where(skiing['WEEKDAY'] == 5, 1, 0)
skiing['SAT'] = np.where((skiing['WEEKDAY'] == 6) | (skiing['WEEKDAY'] == 4), 0.5,skiing['SAT'])
skiing['SUN'] = np.where(skiing['WEEKDAY'] == 6, 1, 0)
skiing['SUN'] = np.where((skiing['WEEKDAY'] == 0) | (skiing['WEEKDAY'] == 5), 0.5,skiing['SUN'])
```

```python
skiing.to_csv('./data/SteamboatData.csv')
```

```python
max1hr = skiing[col_list].max().max()

for col in col_list:
    skiing[col] = skiing[col] / max1hr
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

skiing[['Flux','PrevFlux24','PrevFlux48']] = scaler.fit_transform(skiing[['Flux','PrevFlux24','PrevFlux48']])
skiing.head()
```

```python
maxDepth = skiing['SNWD'].max()
skiing['SNWD'] = skiing['SNWD'] / maxDepth
maxSnow = skiing['SNOW'].max()
skiing['SNOW'] = skiing['SNOW'] / maxSnow
maxT = skiing['TMAX'].max()
minT = skiing['TMIN'].min()
skiing['TMAX'] = (skiing['TMAX']-minT)/(maxT-minT)
skiing['TMIN'] = (skiing['TMIN']-minT)/(maxT-minT)
```

```python
skiing['SNOW24'] = skiing['SNOW'].rolling(window=1, min_periods=1).sum().shift().bfill()
skiing['SNOW48'] = skiing['SNOW'].rolling(window=2, min_periods=1).sum().shift().bfill()
skiing['SNOW7'] = skiing['SNOW'].rolling(window=7, min_periods=1).sum().shift().bfill()
```

```python
skiing = skiing.drop(columns="WEEKDAY")
```
