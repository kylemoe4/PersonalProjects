```python
import numpy as np
import matplotlib.pyplot as plt
import math

# 2)

def Newtons(f, dfdx, x0, eps,xnarr,fnarr,N=20):
    # f = f(x), dfdx = f'(x), x0 = initial x, eps = tolerance
    xn = x0
    while (True): 
        fxn = f(xn)
        xnarr.append(xn)
        fnarr.append(fxn)
        if (N <= 0):
            return xn
        if abs(fxn) < eps:
            return xn
        dfdxn = dfdx(xn)
        if dfdxn == 0:
            return None
        xn = xn - fxn/dfdxn
        N-=1;
        
```

```python
eat = lambda x: x**2-2
deat = lambda x: 2*x
eat2 = lambda x: x**4-4*x**2+4
deat2 = lambda x: 4*x**3-8*x
eatArr = []
eatfArr = []
print(Newtons(eat,deat,10,0.00001,eatArr,eatfArr))
print(eatArr)

eatArr = []
eatfArr = []
print(Newtons(eat2,deat2,10,0.00001,eatArr,eatfArr))
print(eatArr)
```

```python
# 2) a)

f = lambda x: (x-3)*(x-2)
dfdx = lambda x: 2*x-5
```

```python
x = np.linspace(1,4,151)
dx = np.diff(f(x))/0.02

plt.plot(x[:-1], dx,lw=5,label="Diff Approx")
plt.plot(x,dfdx(x),lw=3,label="True df/dx")
plt.legend(loc="upper left")
plt.show()
```

```python
# See how substantially low the error is as the above plot suggests strong 
# similarity between our numerical approximation and the explicit derivative
plt.plot(x[:-1],dfdx(x)[:-1]-dx)
# error is of order 1e-14 so this is a good approximation
```
