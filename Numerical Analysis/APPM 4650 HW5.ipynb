```python
# 2) 

import math
import numpy as np
import matplotlib.pyplot as plt

f = lambda h: math.exp(3*(0.3+h))
fp = 3*math.exp(.9)
print(fp)
hVals = np.linspace(5*10**-7,10**-5,2000)
cent3pt = lambda h: (np.exp(3*(0.3+h))-np.exp(3*(0.3-h)))/(2*h)
```

```python
plt.plot(hVals,cent3pt(hVals))
plt.title("3 Point Centered Difference Formula Error")
plt.ylabel("Error")
plt.xlabel("h Value")
finds = (abs(cent3pt(hVals)-fp) < 3*10**-12)

print("The following are error values at a particular h.")
for i in range(0,len(hVals)):
    if finds[i] == True:
        print(str(hVals[i]) + " -> " + str(abs(cent3pt(hVals[i])-fp)))
```

```python
# 3)

hs = np.linspace(10**-4,10**-2,100)
err = lambda h: h**2+10**-8*2/h
plt.plot(hs,err(hs))
plt.title("Centered Difference Formula Error as function of h")
plt.ylabel("Error")
plt.xlabel("h Value")
```
