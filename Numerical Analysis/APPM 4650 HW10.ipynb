```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 2) d)
f = lambda t: np.cos(t)
y = lambda t: np.sin(t)

def explicit(f,a,b,n,alpha):
    h = (b-a)/n
    w = [alpha]
    w.append(y(h))
    for i in range(2,n+1):
        t = i*h
        w.append(4*w[i-1]-3*w[i-2]-2*h*f(t-2*h))
    return w
print("y(0.1) ~= " +str(explicit(f,0,0.1,10,0)[10]))
```

```python
def implicit(f,a,b,n,alpha):
    h = (b-a)/n
    w = [alpha]
    w.append(y(h))
    for i in range(2,n+1):
        t=i*h
        w.append((1/3)*(4*w[i-1]-w[i-2]+2*h*f(t)))
    return w
print("y(0.1) ~= " +str(implicit(f,0,0.1,10,0)[10]))
```

```python
# 2) e)
errImp = []
act = y(0.1)
nvals = np.logspace(1,4,12)
for i in range(len(nvals)):
    n = int(nvals[i])
    errImp.append(abs(act-implicit(f,0,.1,n,0)[n]))
plt.loglog(nvals,errImp)
```
